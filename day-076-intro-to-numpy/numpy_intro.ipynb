#### 1-Dimensional Arrays (Vectors)

```python
# 1-D array from a list
array_1d = np.array([0, 1.1, 2.2, 3.3, 4.4])
array_1d
```

```python
array_1d.shape
```

```python
# Checking dimension
array_1d.ndim
```

```python
# Accessing values
array_1d[2]
```

#### 2-Dimensional Arrays (Matrices)

```python
array_2d = np.array(
    [
        [1, 2, 3, 9],
        [5, 6, 7, 8],
    ]
)
array_2d
```

```python
display(f"shape: {array_2d.shape}")
display(f"dimension {array_1d.ndim}")
```

```python
# ndarrays have homogenous data types, which can be checked by .dtype
array_2d.dtype
```

```python
print(array_2d)
```

```python
# Accessing: 1st value in bracket is the row (0-indexed), and the column the
# value is in.
array_2d[1, 2]
```

#### N-Dimensional Arrays (Tensors)

**Challenge**: 
* How many dimensions does the array below have? 
* What is its shape (i.e., how many elements are along each axis)?
* Try to access the value `18` in the last line of code.
* Try to retrieve a 1 dimensional vector with the values `[97, 0, 27, 18]`
* Try to retrieve a (3,2) matrix with the values `[[ 0,  4], [ 7,  5], [ 5, 97]]`

*Hint*: You can use the `:` operator just as with Python Lists.

```python
mystery_array = np.array(
    [
        [[0, 1, 2, 3], [4, 5, 6, 7]],
        [[7, 86, 6, 98], [5, 1, 0, 4]],
        [[5, 36, 32, 48], [97, 0, 27, 18]],
    ]
)

# Note all the square brackets!
```

```python
# Challenge 1 (C1) How many dimensions does the array below have?
mystery_array.ndim
```

```python
# C2. What is its shape (i.e., how many elements are along each axis)?
mystery_array.shape
# This means 3 layers, 2 rows, and 4 columns
```

```python
# Let's see the array to confirm
print(mystery_array)
```

```python
# C3. Try to access the value `18` in the last line of code.
# 3rd "layer", 2nd row, 4th column (2, 1, 3 since 0-indexed)
mystery_array[2, 1, 3]
```

```python
print(mystery_array)
```

```python
# C4. Try to retrieve a 1 dimensional vector with the values `[97, 0, 27, 18]`
# 3rd layer, 2nd row (2, 1)
mystery_array[2, 1]
# Or
# mytery_array[2, 1, :]. Since the last argument happens to be the column, we can omit
# it and it will automatically get all columns
```

```python
print(mystery_array)
```

```python
# C5. Try to retrieve a (3,2) matrix with the values `[[ 0,  4], [ 7,  5], [ 5, 97]]`
# All the first column (0) on all layers (:, :)
mystery_array[:, :, 0]
```

# Axes (Axis)

For C3 to C5, I've used the more intuitive way of visualizing the ndarray by using "layers", rows, and columns to refer to the different dimensions. Generally, each dimension can be referred to as nth axis.

To illustrate this, we will solve C3 again but this time I will be referring to the different dimensions as axis.

```python
print(mystery_array)
```

```python
# C3. Try to access the value `18` in the last line of code.
# 18 is:
# in the 3rd element of the 1st axis ("layers")
# in the 2nd element of the 2nd axis (rows)
# in the 4th element of the 3rd axis (columns)
mystery_array[2, 1, 3]
```

# Dimensions

Each hierarchical grouping of data is what defines a new dimension. 

```python
# 1-dimension:
print(array_1d)
# Note how there is only 1 grouping (signified by a pair of brackets)
```

```python
# 2-dimension:
print(array_2d)
# pair of brackets enclosing multiple groups
```

```python
# 3-dimension:
print(mystery_array)
# pair of brackets enclosing groups of groups
```
